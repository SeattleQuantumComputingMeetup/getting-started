- What is a quantum phase? (Google AI Overview) 
  - In quantum computing, a quantum phase is a component of a quantum state's complex amplitude, similar to a phase in wave physics, which is crucial for quantum interference and enabling algorithms like Quantum Phase Estimation (QPE) to extract information from quantum systems. A quantum state is a superposition of basis states, each with an amplitude that includes both magnitude and a phase, denoted as eiΦ. While the magnitude determines the probability of measuring a specific state, the relative phase between different states allows for constructive and destructive interference, which is fundamental for the power of quantum algorithms.
  - relative: The phase difference between two different components of a superposition. It is the relative phase that matters for quantum interference and is therefore crucial for the operation of quantum computers. 
  - global: A phase that is common to all parts of a quantum state. It is not physically observable and does not affect the probabilities of measurement outcomes, so it is often discarded.

  - Error mitigation and error suppression are both error-handling techniques in computing, but they differ in their approach: error suppression is a proactive strategy to prevent errors from occurring or to minimize their probability of happening during computation, while error mitigation is a post-execution process that uses classical post-processing on raw measurement results to reduce or eliminate the impact of errors that have already occurred. #### Quantum states and their representation as state vectors

```python
### create the |+> Bell State

qc = QuantumCircuit(2)

# if we add an x gate here, we would add a phase, and create a different Bell State
# qc.x(0)
qc.h(0)
qc.cx(0, 1)

qc.draw("mpl")
```

```python
from qiskit.quantum_info import Statevector
from qiskit.visualization import plot_bloch_multivector, plot_state_qsphere

sv = Statevector(qc)

print()

display(sv.draw())

display(sv.draw(output='latex', prefix='Statevector: '))

print("Length of the statevector:", len(sv))
print()
print("Probabilities of each element of the statevector:")
print()
print("00:", sv[0]**2)
print("01:", sv[1]**2)
print("10:", sv[2]**2)
print("11:", sv[3]**2)
print()
print("Compare with the probabilities dictionary:")
print()
probs = sv.probabilities_dict()
print(f'Probabilities: {probs}')
print()
print("00:", probs["00"])
print("11:", probs["11"])
print()
```

```python
# Visualize the quantum state of the entire system of 2 qubits

plot_state_qsphere(sv)
```

#### Quantum gates, their representation as matrices, 
and the ways to calculate their effect on quantum states by multiplying the state vector by the gate matrix

```python
### what is the unitary matrix operator for a quantum circuit that generates the |+> Bell State from 2 qubits initially each in the |0> state?
from qiskit.quantum_info import Operator

op = Operator(qc)

unitary_matrix = op.data

print()
print("Unitary Matrix:")
print()
print(unitary_matrix)
print()
```

#### Measurements for single- and multi-qubit systems, 
their outcomes, and the relationship between the amplitudes of a state vector and the probabilities of measurement outcomes for it

```python
### Let's look at the |+> Bell State, if we run it using a perfect simulator, or on a simulator with noise built-in
```

```python
from qiskit.transpiler.preset_passmanagers import generate_preset_pass_manager
from qiskit.visualization import plot_histogram, plot_bloch_multivector
from qiskit_aer import AerSimulator
from qiskit_ibm_runtime import SamplerV2 as Sampler

qc.measure_all()

backend = AerSimulator()

pm = generate_preset_pass_manager(backend=backend, optimization_level=1)
isa_qc = pm.run(qc)

sampler = Sampler(mode=backend)

job = sampler.run([isa_qc], shots=1000)
result = job.result()

counts = result[0].data.meas.get_counts()

plot_histogram(counts)
```

```python
from qiskit_ibm_runtime.fake_provider import FakeTorino

qc.measure_all()

backend = FakeTorino()

pm = generate_preset_pass_manager(backend=backend, optimization_level=1)
isa_qc = pm.run(qc)

sampler = Sampler(mode=backend)

job = sampler.run([isa_qc], shots=1000)
result = job.result()

counts = result[0].data.meas.get_counts()

plot_histogram(counts)
```

#### Dirac notation for quantum states,

- To work with Dirac notation, you use kets |ψ⟩ (state vectors), bras ⟨ψ| (complex conjugates), and bra-ket pairs ⟨ψ|φ⟩ (inner products) to represent quantum states, amplitudes, and probabilities. You can manipulate these with algebraic rules, such as the distributive property for tensor products and the definition of the inner product as a measure of vector alignment. (Google AI Overview)
