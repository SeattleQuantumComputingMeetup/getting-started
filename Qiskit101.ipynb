### Quantum Mechanics and Quantum Hardware

<br>

**Quantum Effects**
- *Superposition* - can create using the Hadamard gate
  
- *Entanglement* - can create using a Hadamard gate on qubit 0 followed by a CNOT gate on qubits 0 and 1

- *Interference* -
  "Quantum gates such as the Hadamard (H) and Phase (S, T, Z) gates create interference by manipulating the phases and amplitudes of qubits in superposition, allowing for constructive and destructive interference. Interference is a key mechanism in quantum algorithms like the Quantum Fourier Transform (QFT), where it amplifies correct answers and cancels out incorrect ones, a principle also seen in Grover's algorithm and other quantum speedups.", GOOG search.
