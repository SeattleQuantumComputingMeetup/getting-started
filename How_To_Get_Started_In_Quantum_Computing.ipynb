### What is Quantum Computing?

- Something to use for problems that we can't address efficiently with classical computers (current thinking)
- Something which makes use of quantum effects: Superposition, Entanglement, and Interference
- A quantum computer program is a definition of a quantum circuit that we run on some number of qubits, multiple times (called shots), in which case we are sampling from a distribution that the circuit defines - in *Qiskit*, we use the Sampler primitive to do this.
  
*Qiskit is an open-source, Python-based, high-performance software stack for quantum computing*, 
- developed by IBM Research and first released in 2017, aka Quantum Information Software Kit
- it provides tools for creating quantum programs and executing them on quantum computers or classical simulators.

Qiskit has another primitive, called the Estimator.
  - the Estimator computes the expectation value of a quantum circuit's state with respect to a given observable
  - we still sample the distribution of a quantum circuit, and we combine that with an Observable (a term from quantum mechanics)

<br>

We want to use quantum computers in places where we have an advantage over classical computers, (graph taken from Mariia Mykhailova's book, [Quantum Programming in Depth](https://www.amazon.com/Quantum-Programming-Depth-Solving-problems/dp/163343690X/ref=sr_1_1?crid=2JP5DYDMVWYCS&dib=eyJ2IjoiMSJ9.RQ1Uiw9sEPanbSwM-Bzm6NbVI12nk0NwBprVyNAmRhA.-HDMjWv0xvehMiFHxVOKCQ_BO4CTyUHepzqiHqeompc&dib_tag=se&keywords=mariia+mykhailova&qid=1758055347&sprefix=mariia+myk%2Caps%2C150&sr=8-1).)

### What is the status of Quantum Computing?

<br>

It's early days of development, but we have come a long way.  

<br>

**MIT Sloan** has a new report out which [captures the state of quantum computing up to 2024](https://mitsloan.mit.edu/ideas-made-to-matter/new-mit-report-captures-state-quantum-computing).

- QPUs performance is improving, but can't yet run large-scale commercial applications
- Quantum technology patents have increased 5-fold since 2024
  - 54% filed by corporations
  - 37% filed by universities
  - China holds 60% of quantum patents in 2024, followed by the US and Japan.
- VC funding for quantum technology reached a new high in 2024
  - 1.6B USD of publicly announced funds went to quantum computing firms
  - 621m USD to quantum software companies
  - Quantum received < 1 % of total venture capital funding worldwide.
- QC buzz in the news, 2024
- demand for quantum skills has nearly tripled since 2018

### What's needed in Quantum Computing?

- Hardware - more accurate qubits, and the ability to scale to a larger number of qubits, necessary controls, and connections
- Software - 70% of the software stack, a scalable quantum error correction method, reported at [IEEE Quantum Week 2025](https://qce.quantum.ieee.org/2025/).
- Algorithms - which outperform classical algorithms
- Applications - what will we use QCs for?

### Topics to Learn Before Starting Your Second Book on Quantum Computing

(Taken from the book, Quantum Programming in Depth: Solving problems with Q# and Qiskit, by Mariia Mykhailova)

- Quantum states and their representation as state vectors
- Quantum gates, their representation as matrices, and the ways to calculate their effect on quantum states by multiplying the state vector by the gate matrix
- Dirac notation for quantum states and performing computations in Dirac notation directly, without reverting to matrix computations
- The main quantum gates: Pauli gates X, Y, and Z, Hadamard gate H, rotation gate Ry, phase shift gates S and T, and CNOT and CCNOT gates
- Controlled and adjoint variants of gates
- Measurements for single- and multi-qubit systems, their outcomes, and the relationship between the amplitudes of a state vector and the probabilities of measurement outcomes for it
